```python
import pandas as pd
import statsmodels.api as sm

# üîÅ Fonction pour tester une variable de contr√¥le
def tester_variable(variable):
    print(f"\n--- Test de la variable : {variable} ---")

    # ‚úÖ Variables de base
    colonnes = ['Valeur fonciere', 'Surface reelle bati', 'Surface terrain',
                'Nombre pieces principales', 'Ann√©e_construction', 'Date mutation',
                'Type local', 'Etiquette_DPE', 'Nature mutation', 'Code postal']

    if variable not in colonnes:
        colonnes.append(variable)

    df_test = df[colonnes].copy()

    # ‚è≥ Extraire ann√©e de mutation
    df_test['Ann√©e mutation'] = pd.to_datetime(df_test['Date mutation'], errors='coerce').dt.year
    df_test['Code postal simplifie'] = df_test['Code postal'].astype(str).str[:2]

    # üßΩ Nettoyage des variables num√©riques
    for col in df_test.columns:
        if col != 'Date mutation' and df_test[col].dtype == 'object':
            df_test[col] = (
                df_test[col]
                .astype(str)
                .str.replace(',', '.', regex=False)
                .str.replace(' ', '', regex=False)
            )
        try:
            df_test[col] = pd.to_numeric(df_test[col], errors='coerce')
        except:
            pass  # certaines colonnes restent cat√©gorielles

    # üéØ Encodage des variables cat√©gorielles
    cat_vars = ['Type local', 'Etiquette_DPE', 'Nature mutation', 'Code postal simplifie']
    if variable in df_test.columns and df_test[variable].dtype == 'object':
        cat_vars.append(variable)

    df_test = pd.get_dummies(df_test, columns=cat_vars, drop_first=True)

    # üß™ Construction de X et y
    X = df_test.drop(columns=['Valeur fonciere', 'Code postal', 'Date mutation'])
    y = df_test['Valeur fonciere']

    # üîç Nettoyage uniquement sur X et y
    colonnes_utiles = X.columns.tolist() + ['y']
    df_reg = pd.concat([X, y.rename('y')], axis=1).dropna(subset=colonnes_utiles)
    X = df_reg.drop(columns='y')
    y = df_reg['y'].values.ravel()

    # üí£ V√©rif de viabilit√©
    if len(X) < X.shape[1] + 1:
        print(f"‚ö†Ô∏è Trop peu d'observations ({len(X)}) pour {X.shape[1]} variables.")
        return

    # ‚úÖ R√©gression
    X = X.astype(float)
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()

    # üìä R√©sultats
    print(f"‚úÖ R¬≤ ajust√© : {round(model.rsquared_adj, 4)}")
    print(f"üìè Nb d'observations : {len(X)}")
    print("üìå Coefficients des DPE :")
    dpe_cols = [col for col in model.params.index if 'Etiquette_DPE' in col]
    for col in dpe_cols:
        coef = round(model.params[col], 2)
        pval = model.pvalues[col]
        signif = "***" if pval < 0.001 else "**" if pval < 0.01 else "*" if pval < 0.05 else ""
        print(f"  {col} : {coef} {signif}")

# üì¶ Variables √† tester
variables_candidates = [
    'Surface Carrez du 1er lot',
    'Nombre de lots',
    'Code type local',
    'P√©riode_construction'
]

# üöÄ Lancer les tests
for var in variables_candidates:
    tester_variable(var)
```

```python
import pandas as pd
import statsmodels.api as sm

# üîÅ Fonction de test
def tester_variable(variable):
    print(f"\n--- Test de la variable : {variable} ---")

    colonnes = ['Valeur fonciere', 'Surface reelle bati', 'Surface terrain',
                'Nombre pieces principales', 'Ann√©e_construction', 'Date mutation',
                'Type local', 'Etiquette_DPE', 'Nature mutation', 'Code postal']
# ‚õî On ne doit pas chercher √† s√©lectionner 'Ann√©e mutation' dans df
base_colonnes = colonnes.copy()
if variable not in base_colonnes:
    base_colonnes.append(variable)

df_test = df[base_colonnes].copy()


    # Extraction ann√©e mutation
df_test['Ann√©e mutation'] = pd.to_datetime(df_test['Date mutation'], errors='coerce').dt.year
df_test['Code postal simplifie'] = df_test['Code postal'].astype(str).str[:2]

    # Nettoyage num√©rique
    for col in df_test.columns:
        if col != 'Date mutation' and df_test[col].dtype == 'object':
            df_test[col] = df_test[col].astype(str).str.replace(',', '.', regex=False).str.replace(' ', '', regex=False)
        try:
            df_test[col] = pd.to_numeric(df_test[col], errors='coerce')
        except:
            pass

    # V√©rifie taux de NaN sur la variable test√©e
    if df_test[variable].isna().mean() > 0.5:
        print(f"‚ùå Trop de NaN dans la variable : {variable}")
        return

    # Encodage des variables cat√©gorielles
    cat_vars = ['Type local', 'Etiquette_DPE', 'Nature mutation', 'Code postal simplifie']
    if variable in df_test.columns and df_test[variable].dtype == 'object':
        cat_vars.append(variable)

    df_test = pd.get_dummies(df_test, columns=cat_vars, drop_first=True)

    # S√©paration X/y
    X = df_test.drop(columns=['Valeur fonciere', 'Code postal', 'Date mutation'])
    y = df_test['Valeur fonciere']

    colonnes_utiles = X.columns.tolist() + ['y']
    df_reg = pd.concat([X, y.rename('y')], axis=1).dropna(subset=colonnes_utiles)
    X = df_reg.drop(columns='y')
    y = df_reg['y'].values.ravel()

    if len(X) < X.shape[1] + 1:
        print(f"‚ö†Ô∏è Trop peu d'observations ({len(X)}) pour {X.shape[1]} variables.")
        return

    X = X.astype(float)
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()

    print(f"‚úÖ R¬≤ ajust√© : {round(model.rsquared_adj, 4)}")
    print(f"üìè Nb d'observations : {len(X)}")
    dpe_cols = [col for col in model.params.index if 'Etiquette_DPE' in col]
    if dpe_cols:
        print("üìå Coefficients des DPE :")
        for col in dpe_cols:
            coef = round(model.params[col], 2)
            pval = model.pvalues[col]
            signif = "***" if pval < 0.001 else "**" if pval < 0.01 else "*" if pval < 0.05 else ""
            print(f"  {col} : {coef} {signif}")
    else:
        print("üìå Pas de variables DPE dans ce mod√®le.")

# üöÄ Variables candidates √† tester
variables_candidates = [
    'Ann√©e mutation',
    'Type de voie',
    'Surface terrain',
    'B/T/Q',
    'Section',
    '1 Articles CGI'
]

# üîÑ Lancement
for var in variables_candidates:
    tester_variable(var)
```
