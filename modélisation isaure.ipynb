```python
import pandas as pd
import statsmodels.api as sm

# 🔁 Fonction pour tester une variable de contrôle
def tester_variable(variable):
    print(f"\n--- Test de la variable : {variable} ---")

    # ✅ Variables de base
    colonnes = ['Valeur fonciere', 'Surface reelle bati', 'Surface terrain',
                'Nombre pieces principales', 'Année_construction', 'Date mutation',
                'Type local', 'Etiquette_DPE', 'Nature mutation', 'Code postal']

    if variable not in colonnes:
        colonnes.append(variable)

    df_test = df[colonnes].copy()

    # ⏳ Extraire année de mutation
    df_test['Année mutation'] = pd.to_datetime(df_test['Date mutation'], errors='coerce').dt.year
    df_test['Code postal simplifie'] = df_test['Code postal'].astype(str).str[:2]

    # 🧽 Nettoyage des variables numériques
    for col in df_test.columns:
        if col != 'Date mutation' and df_test[col].dtype == 'object':
            df_test[col] = (
                df_test[col]
                .astype(str)
                .str.replace(',', '.', regex=False)
                .str.replace(' ', '', regex=False)
            )
        try:
            df_test[col] = pd.to_numeric(df_test[col], errors='coerce')
        except:
            pass  # certaines colonnes restent catégorielles

    # 🎯 Encodage des variables catégorielles
    cat_vars = ['Type local', 'Etiquette_DPE', 'Nature mutation', 'Code postal simplifie']
    if variable in df_test.columns and df_test[variable].dtype == 'object':
        cat_vars.append(variable)

    df_test = pd.get_dummies(df_test, columns=cat_vars, drop_first=True)

    # 🧪 Construction de X et y
    X = df_test.drop(columns=['Valeur fonciere', 'Code postal', 'Date mutation'])
    y = df_test['Valeur fonciere']

    # 🔍 Nettoyage uniquement sur X et y
    colonnes_utiles = X.columns.tolist() + ['y']
    df_reg = pd.concat([X, y.rename('y')], axis=1).dropna(subset=colonnes_utiles)
    X = df_reg.drop(columns='y')
    y = df_reg['y'].values.ravel()

    # 💣 Vérif de viabilité
    if len(X) < X.shape[1] + 1:
        print(f"⚠️ Trop peu d'observations ({len(X)}) pour {X.shape[1]} variables.")
        return

    # ✅ Régression
    X = X.astype(float)
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()

    # 📊 Résultats
    print(f"✅ R² ajusté : {round(model.rsquared_adj, 4)}")
    print(f"📏 Nb d'observations : {len(X)}")
    print("📌 Coefficients des DPE :")
    dpe_cols = [col for col in model.params.index if 'Etiquette_DPE' in col]
    for col in dpe_cols:
        coef = round(model.params[col], 2)
        pval = model.pvalues[col]
        signif = "***" if pval < 0.001 else "**" if pval < 0.01 else "*" if pval < 0.05 else ""
        print(f"  {col} : {coef} {signif}")

# 📦 Variables à tester
variables_candidates = [
    'Surface Carrez du 1er lot',
    'Nombre de lots',
    'Code type local',
    'Période_construction'
]

# 🚀 Lancer les tests
for var in variables_candidates:
    tester_variable(var)
```
